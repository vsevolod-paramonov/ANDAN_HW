### Задача 1 (Парадокс инспектора)

Автобусы отходят от автостанции с 8:00 до 20:00. Первый автобус отходит ровно в 8:00. Затем интервалы между автобусами случайны, независимы и равновероятно равны либо 5-и, либо 10-и минутам. Будем считать, что за минуту на автостанцию приходит ровно один пассажир, и все пассажиры едут ближайшим автобусом.

Проведите $10^4$ экспериментов и с их помощью:

__а) [5 баллов]__ Постройте гистограмму количества автобусов, отошедших от автостанции за сутки. Похоже ли визуально распределение на нормальное?

__б) [10 баллов]__ Инспектор Тимон выбирает равновероятно один из всех автобусов отошедших от автостанции за сутки. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__в) [10 баллов]__ Инспектор Пумба приходит на автостанцию в случайный момент времени, равномерный от 8:00 до 20:00 и садится в первый пришедший автобус. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__г) [10 баллов]__ Как изменятся ответы на эти вопросы, если время между автобусами будет экспоненциально со средним в 10 минут?

__Пояснения:__ первый автобус можно не включать в выборку, Пумбу за пассажира можно не считать. 

<p style='color: red; font-size:30px'><strong>Решение</strong></p>

<p style='font-size:20px'><strong>а)</strong></p>

```python
n_obs = 10**4
vib = np.random.RandomState(111).choice([5, 10], size=(n_obs, 240))

### Гистограмма ###

vib_cusum = np.cumsum(vib, axis=1)
vib[vib_cusum>720] = 0

a = np.count_nonzero(vib, axis=1)

plt.figure(figsize=(10,6))
plt.hist(a, bins=25, density=True, color='orange')


### Проверка на нормальное распределение ###

mn = np.mean(a)
sd = np.std(a)
x = stats.norm(loc=mn, scale=sd)
x_pdf = x.pdf(np.unique(a))


plt.plot(np.unique(a), x_pdf, '--', linewidth=2, color='black')
plt.ylabel('Частота')
plt.xlabel('Количество автобусов, пришедших за сутки')
plt.title("Гистограмма количества автобусов, отошедших от автостанции за сутки ")
plt.show()
```

<p style='font-size:20px; color:green'><strong>Вывод</strong></p>

<p style='font-size:18px'>Построив гистограмму количества автобусов, пришедших в течение дня, можно предположить, что гистограмма похожа на нормальное распределение. Создадим нормально распределенную случайную величину с посчитанным матожиданием и дисперсии для выборки. Черной линией на графике обозначено нормальное распределение. Поскольку графики почти идентичны можно сделать вывод, что количество автобусов, пришедших в течение дня, нормально распределено</p>

<br>

<p style='font-size:20px'><strong>б)</strong></p>

```python
np.random.seed(1)

def random_value(x):
    l = np.random.choice(x[x>0], size=1)[0]
    return l


timon = np.apply_along_axis(random_value, 1, vib)

plt.figure(figsize=(10,6))
plt.hist(timon, density=True, bins=5, color='purple')
plt.ylabel('Частота')
plt.xlabel('Количество людей в автобусе')
plt.title("Гистограмма числа пассажиров в автобусе, выбранном Тимоном")
plt.show()

print('Матожидание: ', np.mean(timon))
print('Дисперсия: ', np.std(timon)**2)
```

<p style='font-size:20px; color:green'><strong>Вывод</strong></p>

<p style='font-size:18px'>Взяв выборку из $10^4$ наблюдений, выбираем рандомное значение из каждой симуляции. Исходя из условий задачи, за 1 минуту может прийти ровно 1 пассажир, следовательно, в автобусе может быть либо 5, либо 10 пассажиров, поскольку автобусы приходят с интервалом либо 5, либо 10 минут. Можно сделать предположение, что по ЦПТ сумма экспериментов будет стремиться к среднему числу, а именно 7,5. Построив гистограмму количества людей в автобусе, который выбрал Тимон, равновероятно принимает значения 5 или 10</p>

<br>

<p style='font-size:20px'><strong>в)</strong></p>

```python
np.random.seed(123)

n_obs = 10**4

x = stats.uniform(0,710)
x = np.reshape(x.rvs(n_obs), (n_obs,1))


dif = vib_cumsum - x
dif[dif<0] = 10000000
ind = np.argmin(dif, axis=1)
rez = np.take_along_axis(vib,np.reshape(ind, (n_obs, 1)), axis=1)

plt.figure(figsize=(10,6))
plt.hist(rez, color='plum')
plt.ylabel('Частота')
plt.xlabel('Количество людей в автобусе')
plt.title("Гистограмма числа пассажиров в автобусе, выбранном Тимоном")
plt.show()

print('Матожидание: ', np.mean(rez))
print('Дисперсия: ', np.std(rez)**2)
```

### Задача 2 (Парадокс Хуана Мануэля Родригеса Паррондо)

У Атоса, Портоса и Арамиса по 1000 франков. Атос постоянно ходит в казино А, где каждый раз выигрывает один франк с вероятностью $0.49$ и проигрывает один франк с вероятностью $0.51$. Портос ходит в казино Б, где ситуация интереснее :) 

- Если богатство посетителя кратно трём, то посетитель выигрывает франк с вероятностью $0.09$ и проигрывает один франк с вероятностью $0.91$
- Если богатство посетителя не кратно трём, то посетитель выигрывает франк с вероятностью $0.74$ и проигрывает один франк с вероятностью $0.26$

Арамис каждый раз выбирает казино А или казино Б равновероятно. Проведите $10^4$ симуляций эволюции благосостояния Атоса, Портоса и Арамиса.Каждая симуляция предусматривает $1000$ посещений казино. Все три игрока все деньги носят с собой и ни на что не тратят :)

__а) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Атоса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__б) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Портоса, по горизонтали —  номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__в) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Арамиса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.
    

<p style='color: red; font-size:30px'><strong>Решение</strong></p>

<p style='font-size:20px'><strong>а)</strong></p>

```python
n_obs = 10**4


np.random.seed(121)

simulations = np.cumsum(np.random.choice([1,-1], p=[0.49, 0.51], size = (n_obs, 1000)), axis=1) + 1000
choices = simulations[np.random.choice(simulations.shape[0], size=10, replace=False)]
plt.figure(figsize=(10,6))
np.apply_along_axis(plt.plot, 1, choices)


f = np.apply_along_axis(np.mean, 0, simulations)

plt.plot(np.arange(1000),f, color='black', linewidth=2)
plt.xlabel("Номер посещения казино")
plt.ylabel('Богатство')
plt.title('Траектории изменений богатства Атоса')
plt.show()
```

<p style='font-size:20px; color:green'><strong>Вывод</strong></p>

<p style='font-size:18px'>Так как вероятность выигрыша меньше, чем проигрыша, следовательно, при достаточно большом количестве игр капитал Атоса будет уменьшаться, что и видно на графике (черная линия - среднее значение капитала для каждой игры по 10 случайно выбранным симуляциям)</p>

<br>

<p style='font-size:20px'><strong>б)</strong></p>

```python
np.random.seed(15)

n_obs = 10**4


def imba(x):
    bank = 1000
    mas = np.zeros(1000)
    for i in range(1000):
        if bank%3 == 0:
            x = np.random.choice([1,-1], p=[0.09, 0.91])
            mas[i] = x
            bank += x
        else:
            x = np.random.choice([1,-1], p=[0.74, 0.26])
            mas[i] = x
            bank += x       
    return mas

matrix = np.full((n_obs,1000), 0)
matrix = np.apply_along_axis(imba, 1, matrix)
matrix = np.cumsum(matrix, axis=1) + 1000

choices = matrix[np.random.choice(matrix.shape[0], size=10, replace=False)]

plt.figure(figsize=(10,6))
np.apply_along_axis(plt.plot, 1, choices)
mean = np.apply_along_axis(np.mean, 0, matrix)

plt.plot(np.arange(1, 1001), mean, color='black', linewidth=2)
plt.xlabel("Номер посещения казино")
plt.ylabel('Богатство')
plt.title('Траектории изменений богатства Портоса')
plt.show()
```

<p style='font-size:20px; color:green'><strong>Вывод</strong></p>

<p style='font-size:18px'>Каждый раз, когда капитал Портоса становится кратным 3, Портос с большой вероятностью проигрывает 1 франк, возвращаясь в предыдущую позицию и вероятнее всего такое будет продолжаться до тех пор, пока при богатстве не кратном 3 он не начнет проигрывать. Поэтому на графике можно наблюдать огромное количество прямых линий в траектории изменения богатства</p>

<br>

<p style='font-size:20px'><strong>в)</strong></p>

```python
np.random.seed(33)

n_obs = 10**4

def pp(x):
    bank = 1000
    mas = np.zeros(1000)
    for i in range(len(x)):
        if x[i] == 'A':
            rv = np.random.choice([1,-1], p=[0.49, 0.51])
            bank += rv
            mas[i] = rv
        elif x[i] == 'B' and bank%3 == 0:
            rv = np.random.choice([1,-1], p=[0.09, 0.91])
            bank += rv
            mas[i] = rv
        elif x[i] == 'B' and bank%3 != 0:
            rv = np.random.choice([1,-1], p=[0.74, 0.26])
            bank += rv
            mas[i] = rv
    return mas



strat = np.random.choice(['A','B'], size=(n_obs, 1000))
strat = np.cumsum(np.apply_along_axis(pp, 1, strat), axis = 1) + 1000

choices = strat[np.random.choice(strat.shape[0], size=10, replace=False)]

plt.figure(figsize=(10,6))
np.apply_along_axis(plt.plot, 1, choices)

mean = np.apply_along_axis(np.mean, 0, strat)

plt.plot(np.arange(1, 1001), mean, color='black', linewidth=2)
plt.xlabel("Номер посещения казино")
plt.ylabel('Богатство')
plt.title('Траектории изменений богатства Арамиса')
plt.show()
```

<p style='font-size:20px; color:green'><strong>Вывод</strong></p>

<p style='font-size:18px'>По парадоксу Парондо , если игрок будет чередовать заведомо проигрышные игры, то в результате он получать прибыль, как и получилось на данном графике. Видно, что среднее значение капитала для каждой игры по 10 случайным стратегиям (черная линия) растет с каждой новой игрой</p>

### Задача 3 (Парадокс Берксона)

Предположим, что результаты ЕГЭ школьников по русскому и математике независимы и хорошо аппроксимируются 
нормальным распределением с ожиданием 60 и стандартным отклонением 10. 

УШЭ (Урюпинская Школа Экономики) ныне престижна и забирает себе всех школьников набравших более $n$ баллов в сумме по русскому и математике. 

Для каждого $n$ от $80$ до $160$ с шагом $5$ случайно создайте $10^4$ школьников и поделите их на прошедших и не прошедших в УШЭ.


__а) [5 баллов]__ Постройте график числа прошедших в УШЭ в зависимости от $n$. 

__б) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди прошедших в УШЭ.

__в) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди не прошедших в УШЭ.

<h1 style='color: red'> Решение </h1>

<p style='font-size:20px'><strong>а)</strong></p>

```python
np.random.seed(911)


n_obs = 10**4
norm = stats.norm(loc=60, scale=10)
rus = norm.rvs((17,n_obs))
math = norm.rvs((17,n_obs))
result = math + rus

check = np.full((n_obs, 17), np.arange(80,165,5))

df = pd.DataFrame(math.ravel(), columns=['MATH'])
df['RUS'] = rus.ravel()
df['SUM'] = df['RUS'] + df['MATH']
df['N'] = check.T.ravel()
df['CHECK'] = df['SUM'] > df['N']

passed = df[df['CHECK'] == True]

plt.figure(figsize=(10,6))
plt.plot(passed.groupby('N').agg('count')['MATH'], color='brown')
plt.xlabel("Значение n")
plt.ylabel('Кол-во прошедших')
plt.title('График числа прошедших в УШЭ в зависимости от n')
plt.show()
```

<p style='font-size:20px; color:green'><strong>Вывод</strong></p>

<p style='font-size:18px'>Поскольку баллы по русскому и математике распределены нормально со средним значением 60 баллов, можно сделать вывод, что по мере увеличения n количество прошедших школьников будет уменьшаться, поскольку сумма баллов, большая чем n, будет встречаться реже</p>

<br>

<p style='font-size:20px'><strong>б)</strong></p>

```python
passed_corr = passed.groupby('N').corr().unstack().iloc[:,1]
passed_avg_m = passed.groupby('N')['MATH'].mean()
passed_avg_r = passed.groupby('N')['RUS'].mean()

figure, axis = plt.subplots(1, 3, figsize=(20,6))


axis[0].plot(passed_corr, color='orange')
axis[0].set(title = 'График выборочной корреляции баллов среди прошедших', xlabel='Значение n', ylabel='Значение корреляции')

axis[1].plot(passed_avg_m, color='slateblue')
axis[1].set(title = 'Средние баллы по математике среди прошедших', xlabel='Значение n', ylabel='Кол-во баллов')

axis[2].plot(passed_avg_r, color='olivedrab')
axis[2].set(title = 'Средние баллы по русскому среди прошедших', xlabel='Значение n', ylabel='Кол-во баллов')

plt.show()
```

<p style='font-size:20px; color:green'><strong>Вывод</strong></p>

<p style='font-size:18px'>Посольку баллы по русскому и математике независимы, можно сделать предположение, что корреляция между ними будет равняться 0, но по парадоксу Берксона между ними возникает условная зависимость, именно поэтому на грифке мы видим не нулевую корреляцию. Исходя из растущих графиков средних баллов по предметам, стоит ожидать положительную и растущую корреляцию, но поскольку баллы условно связаны 3 признаком (прошедшие / не прошедшие в УШЭ), то по парадоксу Берксона мы можем наблюдать наоборот падающую корреляцию</p>

<br>

<p style='font-size:20px'><strong>в)</strong></p>

```python
not_passed = df[df['CHECK'] == False]
not_passed_corr = not_passed.groupby('N').corr().unstack().iloc[:,1]

plt.plot(not_passed_corr, color='mediumslateblue')
plt.xlabel("Значение n")
plt.ylabel('Значение корреляции')
plt.title('График выборочной корреляции баллов среди не прошедших')
plt.show()
```

<p style='font-size:20px; color:green'><strong>Вывод</strong></p>

<p style='font-size:18px'>Проводя аналогию с предыдущим пунктом, теперь рассматривается противоположное событие переменной-коллайдера (не прошедшие ученики), поэтому и корреляция на данном графике противоположно направленная предыдущему результату</p>
